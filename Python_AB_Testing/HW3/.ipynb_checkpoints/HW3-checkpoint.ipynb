```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
movie_data = pd.read_csv("HW-3-a.csv")
experiment_data = pd.read_csv("HW-3-b.csv")
```

```python
#Question 1
# Analyzing the distributions of 'n_lease' and 'order'

unique_movies = movie_data['id'].nunique()
print("The number of unique movies in the dataset:\t", unique_movies)

# Distribution of 'n_lease'
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(experiment_data['n_lease'], kde=True)
plt.title('Distribution of n_lease')
plt.xlabel('Number of Leases')
plt.ylabel('Frequency')

# Distribution of 'order'
plt.subplot(1, 2, 2)
sns.histplot(experiment_data['order'], kde=True)
plt.title('Distribution of order')
plt.xlabel('Average Slot Order')
plt.ylabel('Frequency')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Convert 'start_date' to datetime format and extract year
experiment_data['start_date'] = pd.to_datetime(experiment_data['start_date']).dt.year

# One-hot encoding of categorical variables
data_b_encoded = pd.get_dummies(experiment_data, columns=['mac_type', 'movie_group'])

# Extracting the column names for one-hot encoded variables
encoded_columns = data_b_encoded.columns.drop(['n_lease', 'price', 'order', 'start_date'])

# Constructing the model formula
# Including all one-hot encoded variables
formula = 'n_lease ~ price * order + start_date + ' + ' + '.join(encoded_columns)

# Building the regression model with the corrected formula
model = smf.ols(formula, data=data_b_encoded).fit()

# Displaying the summary of the regression results
print(model.summary())
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# To examine how the price elasticity of demand changes with the slot on the TV screen,
# we need to interact the 'price' variable with the 'order' variable.

# We will use a linear regression model with an interaction term between 'price' and 'order'.
# The dependent variable will be 'n_lease'.

# Building the model
model = smf.ols('n_lease ~ price * order', data=experiment_data).fit()

# Displaying the summary of the regression results
model.summary()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Add a column for the logarithm of price
experiment_data['log_price'] = np.log(experiment_data['price'])

# Round the slot values to the nearest integer
experiment_data['rounded_order'] = experiment_data['order'].round().astype(int)

# Initialize a dictionary to store the models
model_summaries = {}

# Define the maximum number of slots to consider in the "left" bundle
max_slots = max(experiment_data['rounded_order'])

# Iterate over the range of slots to create different bundles
for i in range(2, max_slots + 1):
    # Create a binary variable indicating whether a movie is in one of the left slots
    experiment_data['left_slots'] = experiment_data['rounded_order'].apply(lambda x: 1 if x <= i else 0)

    # Build the model with the interaction between left_slots and log_price
    model = smf.ols('n_lease ~ log_price * left_slots', data=experiment_data).fit()

    # Store the model summary
    model_summaries[i] = model.summary()

# Writing the model summaries to a file
file_path = 'model_summaries_bundled_slots.txt'
with open(file_path, 'w') as file:
    for slots, summary in model_summaries.items():
        file.write(f"Model Summary for Bundle Up to Slot {slots}:\n")
        file.write(summary.as_text())
        file.write("\n\n")  # Adding extra lines for readability
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Add a column for the logarithm of price
experiment_data['log_price'] = np.log(experiment_data['price'])

# Round the slot values to the nearest integer
experiment_data['rounded_order'] = experiment_data['order'].round().astype(int)

# One-hot encode categorical variables for fixed effects
experiment_data_encoded = pd.get_dummies(experiment_data, columns=['movie_group', 'mac_type'])

# Initialize a dictionary to store the models
model_summaries = {}

# Define the maximum number of slots to consider in the "left" bundle
max_slots = max(experiment_data_encoded['rounded_order'])

# Extracting the names of the encoded columns for fixed effects
fixed_effects_columns = experiment_data_encoded.columns.drop(['n_lease', 'price', 'log_price', 'order', 'rounded_order', 'start_date', 'packagename'])

# Iterate over the range of slots to create different bundles
for i in range(2, max_slots + 1):
    # Create a binary variable indicating whether a movie is in one of the left slots
    experiment_data_encoded['left_slots'] = experiment_data_encoded['rounded_order'].apply(lambda x: 1 if x <= i else 0)

    # Construct the model formula including fixed effects and the interaction term
    formula = 'n_lease ~ log_price * left_slots + start_date + ' + ' + '.join(fixed_effects_columns)

    # Build the model with the interaction between left_slots and log_price
    model = smf.ols(formula, data=experiment_data_encoded).fit(cov_type='cluster', cov_kwds={'groups': experiment_data_encoded['packagename']})

    # Store the model summary
    model_summaries[i] = model.summary()

# Writing the model summaries to a file
file_path = 'model_summaries_bundled_slots_fixed_effects.txt'
with open(file_path, 'w') as file:
    for slots, summary in model_summaries.items():
        file.write(f"Model Summary for Bundle Up to Slot {slots}:\n")
        file.write(summary.as_text())
        file.write("\n\n")  # Adding extra lines for readability
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Convert 'start_date' to datetime format and extract year
experiment_data['start_date'] = pd.to_datetime(experiment_data['start_date']).dt.year

# One-hot encoding of categorical variables
experiment_data_encoded = pd.get_dummies(experiment_data, columns=['mac_type', 'movie_group'])

# Adding a new column for the logarithm of price
experiment_data_encoded['log_price'] = np.log(experiment_data_encoded['price'])

# Extracting the column names for one-hot encoded variables
encoded_columns = experiment_data_encoded.columns.drop(['n_lease', 'price', 'order', 'start_date', 'log_price'])

# Constructing the model formula with log(price) instead of price
formula_log_price = 'n_lease ~ log_price * order + start_date + ' + ' + '.join(encoded_columns)

# Building the regression model with the new formula
model_log_price = smf.ols(formula_log_price, data=experiment_data_encoded).fit()

# Displaying the summary of the regression results with log(price)
print(model_log_price.summary())
```
